Non-negative Least Squares Demo

```python
import pandas as pd
import numpy as np
from scipy.optimize import nnls
import os, glob

path = r'C:\Users\donolan\OneDrive - Research Triangle Institute\Documents\Dan_ProjectWriting\2023_Dan_STEPS\Ce-MOF Study\DATA\FTIR\csv files\\'

header=[]
for file in os.listdir(path):
    header.append(file)

files = glob.glob(path + '*.csv')
df = pd.concat(
    (
        pd.read_csv(i, sep=',', skiprows=2, usecols=[1],engine='python')          #add appropriate formatting to the read_csv command
                for i in files
    ), 
    axis=1
)
df.columns=header
df.drop(df.tail(30).index,inplace=True)

display(df)
```

```python
a1 = df['A-629-145.csv'].to_list()
a2 = df['SiO2.csv'].to_list()
b1 = df['A-629-155_6.csv'].T.to_list()

A = np.array([a1,a2])
b = np.array(b1)
x, rnorm = nnls(np.transpose(A),b)

print('content of 145 and SiO2 in 155/6 is', x, ', respectively')
print(rnorm)
```
